```python
import pandas as pd
```

```python
# 3. Creating DataFrames
data = {'Name': ['Alice', 'Bob'], 'Age': [25, 30]}
df = pd.DataFrame(data)
```

```python
# 4. Inspecting Data
df.head()       
df.tail()       
df.info()       
df.describe()   
```

```python
# 5. Selecting Data
df['Name']
df[['Name', 'Age']]
df.loc[0]
df.iloc[1]
```

```python
# 6. Filtering Data
df[df['Age'] > 25]  
```

```python
# 7. Adding/Removing Columns
df['Gender'] = ['F', 'M']  
df.drop('Age', axis=1)      # FIXED: Removed inplace=True to preserve 'Age' for step 9
```

```python
# 8. Handling Missing Data
df.isnull()
df.dropna()
df.fillna(0)
```

```python
# 9. Grouping and Aggregating
# WORKS PERFECTLY NOW
df.groupby('Gender')['Age'].mean()
```

```python
# 10. Merging & Joining
df1 = pd.DataFrame({'key': ['A', 'B'], 'Value1': [1, 2]})
df2 = pd.DataFrame({'key': ['A', 'B'], 'Value2': [3, 4]})

pd.merge(df1, df2, on='key')
df1.join(df2, lsuffix='_left', rsuffix='_right') # FIXED: Added suffixes
```
